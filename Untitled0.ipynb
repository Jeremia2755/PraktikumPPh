```python
%%bigquery
-- 1. Laba Kena Pajak dan Pajak Terutang per Tahun
WITH LabaKenaPajak AS (
  SELECT
    tk.tahun,
    tk.pendapatan,
    tk.beban_operasional,
    tk.penyusutan,
    tk.skenario,
    (tk.pendapatan - tk.beban_operasional - tk.penyusutan) AS laba_kena_pajak,
    kf.tax_rate,
    kf.tax_holiday_awal,
    kf.tax_holiday_akhir,
    CASE
      WHEN tk.tahun BETWEEN kf.tax_holiday_awal AND kf.tax_holiday_akhir THEN 0
      ELSE (tk.pendapatan - tk.beban_operasional - tk.penyusutan) * kf.tax_rate
    END AS pajak_terutang,
    CASE
      WHEN tk.tahun BETWEEN kf.tax_holiday_awal AND kf.tax_holiday_akhir THEN 'Ya'
      ELSE 'Tidak'
    END AS dalam_tax_holiday
  FROM main-buffer-441808-u4.PrakPPhbadan.transaksi_keuangan tk
  JOIN main-buffer-441808-u4.PrakPPhbadan.kebijakan_fiska kf
    ON tk.tahun = kf.tahun
)
SELECT
  tahun,
  pendapatan,
  beban_operasional,
  penyusutan,
  skenario,
  laba_kena_pajak,
  tax_rate AS tarif_pajak,
  dalam_tax_holiday,
  pajak_terutang,
  (laba_kena_pajak - pajak_terutang) AS laba_setelah_pajak
FROM LabaKenaPajak
ORDER BY tahun;

-- 2. Analisis Penyusutan Tahunan per Aset
WITH PenyusutanTahunan AS (
  SELECT
    aset_id,
    kategori,
    nilai_perolehan,
    umur_ekonomis,
    metode,
    CASE
      WHEN metode = 'Garis Lurus' THEN nilai_perolehan / umur_ekonomis
      WHEN metode = 'Saldo Menurun' THEN nilai_perolehan * (2.0 / umur_ekonomis) -- Asumsi Double Declining Balance
      ELSE 0
    END AS penyusutan_tahunan
  FROM main-buffer-441808-u4.PrakPPhbadan.aset_tetap
)
SELECT
  aset_id,
  kategori,
  nilai_perolehan,
  umur_ekonomis,
  metode,
  penyusutan_tahunan,
  CASE
    WHEN metode = 'Garis Lurus' THEN 'Konstan setiap tahun'
    WHEN metode = 'Saldo Menurun' THEN 'Menurun setiap tahun'
    ELSE 'Tidak diketahui'
  END AS catatan_metode
FROM PenyusutanTahunan
ORDER BY nilai_perolehan DESC;

-- 3. Ringkasan Kebijakan Fiskal dan Dampaknya
SELECT
  kf.tahun,
  kf.tax_rate,
  kf.tax_holiday_awal,
  kf.tax_holiday_akhir,
  COUNT(DISTINCT tk.tahun) AS jumlah_tahun_transaksi,
  SUM(tk.pendapatan) AS total_pendapatan,
  SUM(tk.beban_operasional) AS total_beban_operasional,
  SUM(tk.penyusutan) AS total_penyusutan
FROM main-buffer-441808-u4.PrakPPhbadan.kebijakan_fiska kf
LEFT JOIN main-buffer-441808-u4.PrakPPhbadan.transaksi_keuangan tk
  ON kf.tahun = tk.tahun
GROUP BY
  kf.tahun,
  kf.tax_rate,
  kf.tax_holiday_awal,
  kf.tax_holiday_akhir
ORDER BY kf.tahun;
    tk.tahun,
    tk.pendapatan,
    tk.beban_operasional,
    tk.penyusutan,
    tk.skenario,
    (tk.pendapatan - tk.beban_operasional - tk.penyusutan) AS laba_kena_pajak,
    kf.tax_rate,
    kf.tax_holiday_awal,
    kf.tax_holiday_akhir,
    CASE
      WHEN tk.tahun BETWEEN kf.tax_holiday_awal AND kf.tax_holiday_akhir THEN 0
      ELSE (tk.pendapatan - tk.beban_operasional - tk.penyusutan) * kf.tax_rate
    END AS pajak_terutang,
    CASE
      WHEN tk.tahun BETWEEN kf.tax_holiday_awal AND kf.tax_holiday_akhir THEN 'Ya'
      ELSE 'Tidak'
    END AS dalam_tax_holiday
  FROM main-buffer-441808-u4.PrakPPhbadan.transaksi_keuangan tk
  JOIN main-buffer-441808-u4.PrakPPhbadan.kebijakan_fiska kf
    ON tk.tahun = kf.tahun
)
SELECT
  tahun,
  pendapatan,
  beban_operasional,
  penyusutan,
  skenario,
  laba_kena_pajak,
  tax_rate AS tarif_pajak,
  dalam_tax_holiday,
  pajak_terutang,
  (laba_kena_pajak - pajak_terutang) AS laba_setelah_pajak
FROM LabaKenaPajak
ORDER BY tahun;

-- 2. Analisis Penyusutan Tahunan per Aset
WITH PenyusutanTahunan AS (
  SELECT
    aset_id,
    kategori,
    nilai_perolehan,
    umur_ekonomis,
    metode,
    CASE
      WHEN metode = 'Garis Lurus' THEN nilai_perolehan / umur_ekonomis
      WHEN metode = 'Saldo Menurun' THEN nilai_perolehan * (2.0 / umur_ekonomis) -- Asumsi Double Declining Balance
      ELSE 0
    END AS penyusutan_tahunan
  FROM main-buffer-441808-u4.PrakPPhbadan.aset_tetap
)
SELECT
  aset_id,
  kategori,
  nilai_perolehan,
  umur_ekonomis,
  metode,
  penyusutan_tahunan,
  CASE
    WHEN metode = 'Garis Lurus' THEN 'Konstan setiap tahun'
    WHEN metode = 'Saldo Menurun' THEN 'Menurun setiap tahun'
    ELSE 'Tidak diketahui'
  END AS catatan_metode
FROM PenyusutanTahunan
ORDER BY nilai_perolehan DESC;

-- 3. Ringkasan Kebijakan Fiskal dan Dampaknya
SELECT
  kf.tahun,
  kf.tax_rate,
  kf.tax_holiday_awal,
  kf.tax_holiday_akhir,
  COUNT(DISTINCT tk.tahun) AS jumlah_tahun_transaksi,
  SUM(tk.pendapatan) AS total_pendapatan,
  SUM(tk.beban_operasional) AS total_beban_operasional,
  SUM(tk.penyusutan) AS total_penyusutan
FROM main-buffer-441808-u4.PrakPPhbadan.kebijakan_fiska kf
LEFT JOIN main-buffer-441808-u4.PrakPPhbadan.transaksi_keuangan tk
  ON kf.tahun = tk.tahun
GROUP BY
  kf.tahun,
  kf.tax_rate,
  kf.tax_holiday_awal,
  kf.tax_holiday_akhir
ORDER BY kf.tahun;
```

## Data cleaning

### Subtask:
Clean the three datasets: `df_transaksi`, `df_aset`, and `df_kebijakan`.


**Reasoning**:
Check data types, handle missing values (though none were found in the initial exploration, double-check is necessary), identify and treat outliers, and ensure consistency across the three datasets.  This involves data type conversions if needed, outlier handling using winsorization, and cross-referencing shared columns (like 'tahun') for consistency.



```python
# Check and correct data types
for col in ['pendapatan', 'beban_operasional', 'penyusutan']:
    if df_transaksi[col].dtype != 'float64':
        df_transaksi[col] = pd.to_numeric(df_transaksi[col], errors='coerce')

# Double-check for missing values and impute if necessary
for df in [df_transaksi, df_aset, df_kebijakan]:
    for col in df.columns:
      if df[col].isnull().any():
        if df[col].dtype in ['int64', 'float64']:
          df[col].fillna(df[col].mean(), inplace=True) # Impute with mean for numerical
        else:
          df[col].fillna(df[col].mode()[0], inplace=True) # Impute with mode for categorical

# Outlier treatment using winsorization (example for 'pendapatan')
for col in ['pendapatan', 'beban_operasional', 'penyusutan']:
    df_transaksi[col] = df_transaksi[col].clip(lower=df_transaksi[col].quantile(0.05), upper=df_transaksi[col].quantile(0.95))

# Ensure consistency in 'tahun' column across dataframes
# Check if 'tahun' columns exist in all dataframes
if {'tahun'}.issubset(df_transaksi.columns) and {'tahun'}.issubset(df_kebijakan.columns):
    # Find common years
    common_years = set(df_transaksi['tahun']) & set(df_kebijakan['tahun'])
    # Filter dataframes based on common years
    df_transaksi = df_transaksi[df_transaksi['tahun'].isin(common_years)]
    df_kebijakan = df_kebijakan[df_kebijakan['tahun'].isin(common_years)]
    print("Dataframes filtered to include only common years.")
else:
    print("The 'tahun' column is not present in all dataframes.")

# Display info of dataframes to check changes
display(df_transaksi.info())
display(df_aset.info())
display(df_kebijakan.info())
```

```python
# Merge df_transaksi and df_kebijakan
try:
    df_merged = pd.merge(df_transaksi, df_kebijakan, on='tahun', how='inner')
    print("Successfully merged df_transaksi and df_kebijakan on 'tahun'")
except KeyError:
    print("Error: 'tahun' column not found in one or both dataframes. Skipping merge.")
    df_merged = df_transaksi.copy()

# Feature Engineering: Calculate Profit and Profit Margin
df_merged['Profit'] = df_merged['pendapatan'] - df_merged['beban_operasional'] - df_merged['penyusutan']
df_merged['Profit_Margin'] = df_merged['Profit'] / df_merged['pendapatan']

# Data Type Conversion
for col in ['pendapatan', 'beban_operasional', 'penyusutan', 'Profit', 'Profit_Margin']:
    if df_merged[col].dtype != 'float64':
        df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

# Handle inconsistencies (example: check for negative profit margin)
if 'Profit_Margin' in df_merged.columns:
    negative_profit_margin = df_merged[df_merged['Profit_Margin'] < 0]
    if not negative_profit_margin.empty:
        print("Warning: Negative profit margins detected.")
        display(negative_profit_margin)
        # Handle the negative profit margins appropriately (e.g., investigate the cause, replace with 0)
    else:
        print("No inconsistencies found in the merged dataframe.")
else:
    print("'Profit_Margin' column not found in the merged dataframe")

display(df_merged.head())
```
